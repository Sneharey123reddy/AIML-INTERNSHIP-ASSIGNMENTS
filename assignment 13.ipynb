```python
'''
#Problem Explanation
Customer segmentation divides customers into groups based on similar behavior.
We use K-Means clustering to group customers using features like Age, Income, and Spending Score.
'''

✅ Customer Groups (Example)
High Income – High Spending
Premium customers
Important for luxury marketing
High Income – Low Spending
Rich but not active buyers
Need targeted promotions
Low Income – High Spending
Young enthusiastic shoppers
Low Income – Low Spending
Budget customers
Average Income – Average Spending
Normal customers
✅ Insights
Businesses can target each group with different marketing strategies.
High spenders are valuable customers.
Segmentation helps companies increase sales
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("Mall_Customers.csv")

# Select features
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Apply K-Means
kmeans = KMeans(n_clusters=5, random_state=0)
data['Cluster'] = kmeans.fit_predict(X)

# Plot clusters
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'])
plt.title("Customer Segments")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```
