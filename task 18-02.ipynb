```python
import pandas as pd

# Create dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [50000, 60000, 55000, 70000, 65000, 72000],
    "Marketing Spend": [5000, 7000, 6000, 8000, 7500, 8200],
    "Profit": [15000, 18000, 16000, 20000, 19000, 21000]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Total Revenue
print("Total Revenue:", df["Revenue"].sum())

# Average Profit
print("Average Profit:", df["Profit"].mean())

# Month with Highest Revenue
print("Highest Revenue Month:\n", df.loc[df["Revenue"].idxmax()])

# Month with Lowest Profit
print("Lowest Profit Month:\n", df.loc[df["Profit"].idxmin()])
```

```python
# Add Profit Margin column
df["Profit Margin (%)"] = (df["Profit"] / df["Revenue"]) * 100

print(df)
```

```python
#Step 4: Insights 
'''
Revenue shows an increasing trend from January to June
Highest revenue was in June (72000)
Profit is also steadily increasing
Marketing spend increases along with revenue
Profit margin remains stable across months
'''
# Step 5: Optional Visualization
import matplotlib.pyplot as plt

# Plot Revenue & Profit
plt.plot(df["Month"], df["Revenue"], label="Revenue")
plt.plot(df["Month"], df["Profit"], label="Profit")

plt.xlabel("Month")
plt.ylabel("Amount")
plt.title("Sales Analysis")
plt.legend()
plt.show()
```

```python
#Step 1: Create Dataset

import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Revenue': [20000, 25000, 30000, 28000, 35000, 40000],
    'Marketing_Spend': [5000, 7000, 8000, 6000, 9000, 10000],
    'Profit': [5000, 8000, 10000, 9000, 12000, 15000]
}

df = pd.DataFrame(data)
print(df)
#Step 2: Line Plot (Revenue Trend)
import matplotlib.pyplot as plt

plt.plot(df['Month'], df['Revenue'])
plt.title("Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
# Step 3: Scatter Plot (Marketing vs Profit)
Python
plt.scatter(df['Marketing_Spend'], df['Profit'])
plt.title("Marketing Spend vs Profit")
plt.xlabel("Marketing Spend")
plt.ylabel("Profit")
plt.show()
#Step 4: Correlation Heatmap
Python
import seaborn as sns
corr = df[['Revenue', 'Marketing_Spend', 'Profit']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```
