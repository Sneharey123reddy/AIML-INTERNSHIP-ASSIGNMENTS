```python
'''
Assignment (24/02/2026)
Assignment Name: Dataset Detective
✅ Problem Explanation
Load a dataset
Display top rows
Find column with highest values
Count missing values
Share insights
'''
import pandas as pd

# Load dataset (example: CSV file)
data = pd.read_csv("50_Startups.csv")

# Display top rows
print("Top 5 Rows:")
print(data.head())

# Find column with highest average value
highest_col = data.mean(numeric_only=True).idxmax()
print("\nColumn with highest average value:", highest_col)

# Count missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Basic dataset info
print("\nDataset Info:")
print(data.describe())
```
