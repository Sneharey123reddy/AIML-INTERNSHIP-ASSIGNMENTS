```python
# Step 1: Create Dataset

import pandas as pd

data = {
    'Engine_Size': [1000, 1200, 1500, 1800, 2000, 2200, 2500, 2700],
    'Mileage': [25, 22, 20, 18, 15, 14, 12, 10],
    'Age': [5, 4, 3, 3, 2, 2, 1, 1],
    'Price': [300000, 350000, 500000, 550000, 650000, 700000, 800000, 850000]
}

df = pd.DataFrame(data)
print(df)
# Step 2: Define Features & Labels
X = df[['Engine_Size', 'Mileage', 'Age']]
y = df['Price']
#Step 3: Train Model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
#Step 4: Predict New Car Price
predicted_price = model.predict([[1500, 20, 3]])
print("Predicted Price:", predicted_price[0])
#Step 5: Print Coefficients
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

print(coefficients)
```

```python
# Step 6: Interpretation
Each coefficient shows how much the price changes when that feature increases by 1 unit.
'''
Example Interpretation:
Engine Size → Positive coefficient
# Larger engine → Higher price
Mileage → Negative coefficient
# More mileage → Lower price
Age → Negative coefficient
# Older car → Lower price
# Which Feature Impacts Price Most?
Check the absolute value of coefficient
print(coefficients.sort_values(by='Coefficient', key=abs, ascending=False))
# The feature with the highest absolute coefficient has the most impact on price.
# Final Comment Example
If Engine Size has highest value:
# “Engine size has the strongest influence on car price.”
If Age/Mileage dominates:
# “Depreciation factors (age/mileage) impact price more significantly.”
'''
```
