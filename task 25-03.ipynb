```python
#import libraries
import random
import nltk
from nltk.tokenize import word_tokenize
from collections import defaultdict
nltk.download('punkt')

# Sample training data
corpus = [
    "I love learning natural language processing",
    "NLP is very interesting and powerful",
    "I enjoy building machine learning models",
    "Text generation is a fascinating task",
    "Artificial intelligence is the future",
    "I love coding in python",
    "Machine learning makes systems smart",
    "Deep learning improves performance",
    "I enjoy solving real world problems",
    "Data science is exciting"
]

# Build bigram model
model = defaultdict(list)

for sentence in corpus:
    tokens = word_tokenize(sentence.lower())
    for i in range(len(tokens)-1):
        model[tokens[i]].append(tokens[i+1])
# Generate Text (10 Words)
def generate_text(start_word, length=10):
    result = [start_word.lower()]
    
    for _ in range(length-1):
        last_word = result[-1]
        next_words = model.get(last_word)
        
        if not next_words:
            break
        
        next_word = random.choice(next_words)
        result.append(next_word)
    
    return " ".join(result)

# User input
user_input = input("Enter a starting word: ")

output = generate_text(user_input, 10)
print("Generated Sentence:\n", output)
```
