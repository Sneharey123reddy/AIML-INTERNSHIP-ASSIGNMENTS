```python
import matplotlib.pyplot as plt

# Sample data
students = ["A", "S", "R", "P", "K"]
marks = [85, 92, 76, 64, 88]

# 1. Bar Chart
plt.figure()
plt.bar(students, marks)
plt.title("Student Marks - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# 2. Pie Chart
plt.figure()
plt.pie(marks, labels=students, autopct='%1.1f%%')
plt.title("Marks Distribution - Pie Chart")
plt.show()

# 3. Histogram
plt.figure()
plt.hist(marks, bins=5)
plt.title("Marks Distribution - Histogram")
plt.xlabel("Marks Range")
plt.ylabel("Frequency")
plt.show()
```

```python
'''
Short Data Story
The bar chart shows that S scored the highest marks, followed by K and A, while P scored the lowest.
The pie chart highlights that S contributes the largest share of total marks, indicating strong performance compared to others.
The histogram shows that most students scored between 70 and 90, meaning the overall class performance is good with only a few low scores.
Overall, the data suggests that the class is performing well, with a majority of students achieving above-average marks and only minor variation in performance.
'''
```
