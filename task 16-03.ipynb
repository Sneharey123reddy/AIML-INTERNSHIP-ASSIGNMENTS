```python
'''
Sentiment Vectorization

Dataset

I love this movie
This movie is terrible
Amazing acting
Worst film ever

Students must:
1. Apply Bag of Words
2. Apply TF-IDF
3. Print matrices
4. Compare results

Extra:

Visualize using pandas dataframe
'''
```

```python
#Step 1: Create Dataset
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
corpus = [
    "I love this movie",
    "This movie is terrible",
    "Amazing acting",
    "Worst film ever"
]
# Step 2: Bag of Words (BoW)n
bow = CountVectorizer()
X_bow = bow.fit_transform(corpus)
bow_df = pd.DataFrame(X_bow.toarray(), columns=bow.get_feature_names_out())
print("Bag of Words Matrix:\n", bow_df)
# Step 3: TF-IDF
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(corpus)
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_df)
# Step 4: Visualization using DataFrame
print("\n--- BoW DataFrame ---")
print(bow_df)
print("\n--- TF-IDF DataFrame ---")
print(tfidf_df)
# Step 5: Comparison
'''
-> Bag of Words (BoW)
Counts word frequency
Simple and easy
Example:
“movie” appears more → higher count
# Does NOT consider importance of words
# TF-IDF
Considers importance of words
Reduces weight of common words (like this, is)
Gives higher value to unique words like:
amazing, worst, terrible
# Key Difference
Feature
BoW
TF-IDF
Value Type
Frequency count
Weighted importance
Common Words
High value
Low value
Rare Words
Normal
High importance
✍️ Final Comment (for submission)
👉 “Bag of Words represents text using word frequency, while TF-IDF improves representation by assigning importance to words. TF-IDF provides better feature representation for sentiment analysis.”
'''
```
