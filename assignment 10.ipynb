```python
Assignment Name: KNN in Real Life
# Problem Explanation
KNN (K-Nearest Neighbors) is used to recommend items by finding users/items that are most similar.
# Real-Life Example (Netflix-like Recommendation)
Using Netflix:
When you watch movies, the system finds users with similar taste
It then recommends movies that those similar users liked
This is based on distance/similarity between users
* Example:
If you like action movies, KNN finds users who also like action → suggests their favorite movies
# How KNN Works (Simple Steps)
Choose a value of K (number of neighbors)
Measure similarity (distance) between users
Pick top K similar users
Recommend items liked by them
#Small Similarity Example
User     A    B   C
Action   5    4   1
Comedy   1    1   5
Drama    2    2   4

# Suppose we want recommendations for User A
Distance(A, B) → very small (similar taste)
Distance(A, C) → large (different taste)
✔ So, User B is nearest neighbor
# Recommendation:
Movies liked by User B will be recommended to User A
# Simple Distance Formula (Euclidean)
This calculates similarity between users.
#Conclusion
KNN recommends based on similarity
No training phase (lazy learning)
Works well for recommendation systems
```
