```python
'''
Assignment Name: Data Doctor
✅ Problem Explanation
Clean a dataset by:
Handling missing values
Removing duplicates
Standardizing text
Explaining importance of cleaning
Python
'''
import pandas as pd

# Load dataset
data = pd.read_csv("student.csv")

# 1. Handle missing values
# Fill numeric columns with mean
data.fillna(data.mean(numeric_only=True), inplace=True)

# Fill text columns with 'Unknown'
data.fillna("Unknown", inplace=True)

# 2. Remove duplicate rows
data.drop_duplicates(inplace=True)

# 3. Standardize text (example: name column)
if 'Name' in data.columns:
    data['Name'] = data['Name'].str.lower().str.strip()

# Display cleaned data
print("Cleaned Data:")
print(data.head())
```

```python
'''
✅ Before vs After (Conceptual)
Missing values → Filled
Duplicate rows → Removed
Inconsistent text → Standardized
✅ Why Data Cleaning Matters
Improves accuracy
Clean data ensures correct results in analysis and machine learning.
Prevents errors
Missing or duplicate data can lead to wrong calculations.
Better model performance
ML models perform better with clean and consistent data.
Consistency in data
Standardized text avoids confusion (e.g., "John", "john", " JOHN").
Saves time in analysis
Clean datasets are easier to work with and visualize.
'''
```
