```python
'''
Project: Loan Approval Predictor

Dataset columns:
Income
Credit Score
Age
Loan Amount
Employment Years

Steps:

Load dataset
Train Decision Tree
Train Random Forest
Compare accuracy
Show feature importance

Save model using pickle
'''
```

```python
#Step 1: Create / Load Dataset

import pandas as pd

data = {
    'Income': [30000, 50000, 70000, 90000, 120000, 40000, 60000, 80000, 100000, 110000],
    'Credit_Score': [600, 650, 700, 750, 800, 620, 680, 720, 770, 790],
    'Age': [25, 30, 35, 40, 45, 28, 33, 38, 42, 44],
    'Loan_Amount': [200000, 250000, 300000, 350000, 400000, 220000, 270000, 320000, 370000, 390000],
    'Employment_Years': [1, 3, 5, 7, 10, 2, 4, 6, 8, 9],
    'Approved': [0, 0, 1, 1, 1, 0, 1, 1, 1, 1]  # Target
}

df = pd.DataFrame(data)
print(df)
# Step 2: Features & Label
X = df.drop('Approved', axis=1)
y = df['Approved']
# Step 3: Train-Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
#Step 4: Train Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)
print("Decision Tree Accuracy:", dt_acc)
#Step 5: Train Random Forest
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_acc)
# Step 6: Compare Accuracy
print(f"Decision Tree Accuracy: {dt_acc:.2f}")
print(f"Random Forest Accuracy: {rf_acc:.2f}")
# Usually:
#Random Forest performs better due to ensemble learning.
# Step 7: Feature Importance (Random Forest)
import pandas as pd
importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

print(importance)
# Step 8: Save Model using Pickle
import pickle
with open('loan_model.pkl', 'wb') as file:
    pickle.dump(rf_model, file)
# Load Model Later
with open('loan_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
# Final Insights
#Random Forest > Decision Tree
# More stable, reduces overfitting
'''
Important Features (example):
Credit Score → Strong impact
Income → High influence
Employment Years → Moderate
Model can be used for real-world loan approval systems
'''
```
