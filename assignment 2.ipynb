```python
'''
Print numbers from 1 to 50:
If divisible by 3 → print Fizz
If divisible by 5 → print Buzz
If divisible by both → print FizzBuzz
Also count how many times each appears.
✅ Python Code (Using Loops & Functions)
Python
'''
def fizz_buzz():
    fizz = 0
    buzz = 0
    fizzbuzz = 0

    for i in range(1, 51):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
            fizzbuzz += 1
        elif i % 3 == 0:
            print("Fizz")
            fizz += 1
        elif i % 5 == 0:
            print("Buzz")
            buzz += 1
        else:
            print(i)

    print("\nCounts:")
    print("Fizz:", fizz)
    print("Buzz:", buzz)
    print("FizzBuzz:", fizzbuzz)

# function call
fizz_buzz()
'''
✅ Expected Counts
Fizz = 13
Buzz = 7
FizzBuzz = 3
✅ Concepts Used
Loop (for)
Conditional statements (if-elif-else)
Functions
'''
```
