```python
from gensim.models import Word2Vec
```

```python
sentences = [
    "college life is full of fun and learning",
    "students attend classes and complete assignments",
    "friends study together in the library",
    "exams are important for every student",
    "teachers guide students for their future",
    "college campus has many facilities",
    "students participate in sports and cultural events",
    "learning new skills is important in college",
    "group projects help in teamwork",
    "students enjoy canteen food during breaks",
    "library is a good place to study",
    "college life teaches discipline and responsibility",
    "students prepare for exams seriously",
    "friends support each other in studies",
    "college memories are unforgettable"
    "sports are essential"
]
```

```python
model=Word2Vec(sentences, vector_size=50, window=2, min_count=1, workers=1)
print("vector for 'sports': ")
print(model.wv["sports"])
print("vector for 'college life': ")
print(model.wv["college life"])
print("\nwords similar to 'food':")
print(model.wv.most_similar("food"))
print("\nwords similar to 'exams':")
print(model.wv.most_similar("exams"))

'''
vector size=50 each word represented by %0 numbers
window=2 models look at 2 words before and after target 
'''
```
