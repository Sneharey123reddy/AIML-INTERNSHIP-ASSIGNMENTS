```python
'''
Assignment (20/02/2026)
Assignment Name: NumPy Speed Test
✅ Problem Explanation
Compare performance of:
Python Lists
NumPy Arrays
Using 1 million numbers and measure execution time.
✅ Python Code
Python
'''
import time
import numpy as np

# Create data
size = 1000000

# Python list
list_data = list(range(size))

# NumPy array
numpy_data = np.array(range(size))

# Test Python list (square operation)
start = time.time()
list_result = [x * x for x in list_data]
end = time.time()
print("List Time:", end - start)

# Test NumPy array (square operation)
start = time.time()
numpy_result = numpy_data * numpy_data
end = time.time()
print("NumPy Time:", end - start)
```
