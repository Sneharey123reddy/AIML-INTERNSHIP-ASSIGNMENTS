```python
1. Google Search (Google)
Uses machine learning to understand your query and rank the most relevant web pages based on past data and user behavior.
2. YouTube Recommendations (YouTube)
ML analyzes your watch history, likes, and time spent to suggest videos similar to your interests.
3. Netflix Recommendations (Netflix)
Uses algorithms like collaborative filtering to recommend shows based on what similar users watch.
4. Google Maps Navigation (Google Maps)
ML predicts traffic conditions and finds the fastest route using real-time and historical data.
5. Voice Assistants (Google Assistant / Siri)
Speech recognition models convert voice to text, and NLP models understand and respond to commands.
6. Email Spam Filter (Gmail)
ML models classify emails as spam or not by learning patterns from millions of labeled emails.
7. Instagram Feed (Instagram)
AI ranks posts based on your interactions, likes, comments, and time spent on content.
8. Online Shopping Suggestions (Amazon / Flipkart)
Recommendation systems suggest products using your browsing and purchase history.
9. Face Unlock in Smartphones
Uses computer vision models to recognize facial features and match them with stored data for authentication.
10. Chatbots (ChatGPT)
Uses natural language processing and large language models to understand and generate human-like responses.
```
