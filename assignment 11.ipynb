```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Dataset (House size in sq.ft vs Price in lakhs)
X = np.array([[500], [800], [1000], [1200], [1500]])
y = np.array([50, 80, 100, 120, 150])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict for new input
new_size = np.array([[1100]])
predicted_price = model.predict(new_size)

print("Predicted Price for 1100 sq.ft:", predicted_price[0])
```

```python

'''
Understanding the Model
The model follows a linear equation:
x = House size
y = Price
Learns relationship from data and predicts new values
'''
```
