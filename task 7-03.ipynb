```python
'''
Assignment 1

Build a Loan Approval Prediction Model

Features:
income
credit_score
loan_amount
employment_years

Target:
loan_approved (0/1)

Steps:
create dataset
train model
evaluate accuracy
predict new customer
```

```python
# Step 1: Create Dataset

import pandas as pd

data = {
    'income': [30000, 50000, 70000, 90000, 120000, 40000, 60000, 80000, 100000, 110000],
    'credit_score': [600, 650, 700, 750, 800, 620, 680, 720, 770, 790],
    'loan_amount': [200000, 250000, 300000, 350000, 400000, 220000, 270000, 320000, 370000, 390000],
    'employment_years': [1, 3, 5, 7, 10, 2, 4, 6, 8, 9],
    'loan_approved': [0, 0, 1, 1, 1, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
print(df)
#Step 2: Define Features & Target
X = df[['income', 'credit_score', 'loan_amount', 'employment_years']]
y = df['loan_approved']
#Step 3: Train-Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
#Step 4: Train Model (Logistic Regression)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
# Step 5: Evaluate Accuracy
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
#Step 6: Predict New Customer
# Example new customer
new_customer = [[75000, 720, 300000, 5]]
prediction = model.predict(new_customer)
print("Loan Approved (1=Yes, 0=No):", prediction[0])
#Final Interpretation
'''
The model uses:
Income
Credit Score
Loan Amount
Employment Years
Accuracy shows how well the model predicts approvals.
✍️ Final Comment (for submission)
👉 “The model predicts loan approval based on financial and employment factors. The achieved accuracy indicates that the model can reasonably classify whether a loan will be approved or not.”
'''
```
